## Execute the SQL Interface

A `sqlite3.Cursor` object is our interface to the database, mostly throught the `execute` method that allows to run any `SQL` query on our database.

```python
conn = sqlite3.connect('database.sqlite')
```

```python
# Interface tu the database
cursor = conn.cursor()
```

```python
type(cursor)
```

## Execute and Fetchall

A shortcut to directly execute the query and gather the results are `execute` and `fetchall` methods:

```python
# fetchall() shows the query's results
cursor.execute("SELECT * FROM sqlite_master").fetchall()
```

## Queries

First of all we can get a list of all the tables saved into the database, this is done by reading the column `name` from the `sqlite_master` metadata table with:

    SELECT name FROM sqlite_master
    
The output of the `execute` and the `fetchall` methods is a tuple inside of the list. 

```python
cursor.execute("SELECT name FROM sqlite_master").fetchall()
```

## Data

```python
sample_data = cursor.execute('SELECT * FROM Iris').fetchall()
sample_data
```

## Limit

It's similar to the .head() method from python

```python
cursor.execute('SELECT * FROM Iris LIMIT 3').fetchall()
```

## Columns

`cursor.description` retrieve a list of tuples describing the columns in a result set.

```python
cursor.description
```

## SQL to Pandas

On `read_sql_query()` we place the **query** and the variable that contains the **cursor** 

```python
iris_data = pd.read_sql_query('SELECT * FROM Iris', conn)
iris_data
```

```python
iris_data.head()
```

```python
iris_data.dtypes
```

```python
# Data extraction from the query 
pd.read_sql_query("SELECT * FROM Iris WHERE Species == 'Iris-setosa'", conn)
```
